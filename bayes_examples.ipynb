- **$P(T \mid \neg P)$**: Probability of a positive test result given not pregnant (false positive rate).
- **$P(\neg T \mid P)$**: Probability of a negative test result given pregnancy (false negative rate).
- **$P(\neg T \mid \neg P)$**: Probability of a negative test result given not pregnant (specificity).

### Bayes' Theorem
Bayes' theorem allows us to update the probability of being pregnant given a positive test result ($P(P \mid T)$):

$$
P(P \mid T) = \frac{P(T \mid P) \cdot P(P)}{P(T)}
$$

Where $P(T)$ is the total probability of a positive test result:

$$
P(T) = P(T \mid P) \cdot P(P) + P(T \mid \neg P) \cdot P(\neg P)
$$

### Practical Example
Let's assume:
- Prior probability of being pregnant ($P(P)$) is 0.1.
- Sensitivity of the test ($P(T \mid P)$) is 0.99.
- False positive rate ($P(T \mid \neg P)$) is 0.05.

First, calculate $P(T)$:

$$
P(T) = (0.99 \cdot 0.1) + (0.05 \cdot 0.9) = 0.099 + 0.045 = 0.144
$$

Next, apply Bayes' theorem:

$$
P(P \mid T) = \frac{0.99 \cdot 0.1}{0.144} = \frac{0.099}{0.144} \approx 0.6875
$$

So, given a positive test result, the updated probability of being pregnant is approximately 68.75%.

**Medical Diagnosis**
* Suppose a certain disease affects 1% of the population ($P(D) = 0.01$). 
* A test for this disease has a 99% sensitivity ($P(T \mid D) = 0.99$)

We want to find the probability that a person has the disease given a positive test result ($P(D \mid T)$).

First, calculate $P(T)$:

\begin{align*} 
    P(T) &= P(T \mid D) \cdot P(D) + P(T \mid \neg D) \cdot P(\neg D) \\ 
    &= (0.99 \cdot 0.01) + (0.05 \cdot 0.99) \\
    &= 0.0099 + 0.0495 
    \\ &= 0.0594
\end{align*}


Next, apply Bayes' theorem:
\begin{align*} 
    P(D \mid T) &= \frac{P(T \mid D) \cdot P(D)}{P(T)} \\
    &= \frac{0.99 \cdot 0.01}{0.0594} \\  
    & \approx 0.167
\end{align*}

So, given a positive test result, the probability that a person has the disease is approximately 16.7%.
